# Outlayer

```python
import matplotlib.pyplot as plt
for i in data.columns:
    if(data[i].dtype in ["int64","float64"]):
        data.boxplot(i)
        plt.show()  
```

```python
columns_to_check = ["OPEN_P", "HIGH_P", "LOW_P", "CLOSE_P", "TURNOVER", "VOLUME"]
outliers_sample = pd.DataFrame()

# تكرار العملية لكل عمود في القائمة
for col in columns_to_check:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    iqr = Q3 - Q1
    max_value = Q3 + (iqr * 1.5)
    min_value = Q1 - (iqr * 1.5)
    print(f"For {col}: max = {max_value}, min = {min_value}")
    outliers = data[(data[col] > max_value) | (data[col] < min_value)]
    sample_outliers = outliers.sample(frac=0.1, random_state=42)

    outliers_sample = pd.concat([outliers_sample, sample_outliers])

    data = data[(data[col] <= max_value) & (data[col] >= min_value)]

# دمج البيانات النظيفة مع العينة المحفوظة من القيم الشاذة
data = pd.concat([data, outliers_sample])

print(f"Shape of outliers sample: {outliers_sample.shape}")
print(f"Shape of final data: {data.shape}")
```

# split X,y

```python
from sklearn.model_selection import train_test_split

X = data.drop(columns=["CLOSE_P"])  
y = data["CLOSE_P"]  
```

# feature selection  by SelectPercentile

```python
from sklearn.feature_selection import SelectPercentile, f_classif 
FeatureSelection = SelectPercentile(score_func=f_classif, percentile=80)
```

```python
from sklearn.feature_selection import f_classif

FeatureSelection = SelectPercentile(score_func=f_classif, percentile=80)
X_selected = FeatureSelection.fit_transform(X, y)
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(X_selected, columns=X.columns[FeatureSelection.get_support()])
```

##### 

# Split The Data Train And Test

```python
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=44, shuffle=True)

print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

# Train Model By Use Gridsearchcv RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

rf_model = RandomForestRegressor(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200], 
    'max_depth': [None, 10, 20],    
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 2, 4],  
}

grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid,
                           scoring='neg_mean_squared_error', cv=5, verbose=2, n_jobs=-1)

grid_search.fit(X_train, y_train)

print("Best Parameters: ", grid_search.best_params_)
best_model = grid_search.best_estimator_
```

# Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
y_pred = best_model.predict(X_test)

# حساب مقاييس التقييم المناسبة
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# طباعة النتائج
print(f"Mean Absolute Error (MAE)",  mae )
print(f"Mean Squared Error (MSE)", mse)
print(f"R-squared Score (R²)", r2)
```

```python
import gradio as gr
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# يفترض أن النموذج تم تدريبه مسبقًا
best_model = RandomForestRegressor(max_depth=None, min_samples_leaf=2, min_samples_split=2, n_estimators=50, random_state=42)
best_model.fit(X_train, y_train)  # تأكد من تدريب النموذج على البيانات الخاصة بك مسبقًا

# دالة التنبؤ
def predict_price(open_p, high_p, low_p, turnover, volume, month):
    new_data = np.array([[open_p, high_p, low_p, turnover, volume, month]])
    new_data_df = pd.DataFrame(new_data, columns=X.columns[FeatureSelection.get_support()])
    prediction = best_model.predict(new_data_df)
    return f"Predicted Price: {prediction[0]:.4f}"

# إنشاء واجهة Gradio
iface = gr.Interface(
    fn=predict_price,
    inputs=[
        gr.Textbox(label="Opening Price"),
        gr.Textbox(label="Highest Price"),
        gr.Textbox(label="Lowest Price"),
        gr.Textbox(label="Turnover"),
        gr.Textbox(label="Volume"),
        gr.Dropdown(choices=list(range(1, 13)), label="Month")
    ],
    outputs=gr.Textbox(label="Prediction Result"),
    title="Stock Price Prediction",
    description="Enter stock details to get predicted price."
)

iface.launch()
```
